# Day and Night Image Classifier
---

The day/night image dataset consists of 200 RGB color images in two categories: day and night. There are equal numbers of each example: 100 day images and 100 night images.

We'd like to build a classifier that can accurately label these images as day or night, and that relies on finding distinguishing features between the two types of images!

*Note: All images come from the [AMOS dataset](http://cs.uky.edu/~jacobs/datasets/amos/) (Archive of Many Outdoor Scenes).*

## Training and Testing Data
The 200 day/night images are separated into training and testing datasets. 

* 60% of these images are training images, we would use to create a classifier.
* 40% are test images, which will we use to test the accuracy of the classifier.

First, we set some variables to keep track of some where our images are stored:

    image_dir_test: the directory where our test image data is stored

````python
# Image data directories
training_dir = "day_night_images/training/"
test_dir = "day_night_images/test/"
````

### 1.1 Load the datasets

We'll use the **load_dataset** funtion in the helper function to load the training day/night images and store all of them in a variable, `IMAGE_LIST`. This list contains the images and their associated label ("day" or "night"). 

For example, the first image-label pair in `IMAGE_LIST` can be accessed by index: 
``` IMAGE_LIST[0][:]```.


````python
def load_dataset(image_dir):
    
    im_list = []
    image_types = ["day", "night"]
    
    
    for im_type in image_types:
        
        # Iterate through each image file in each image_type folder
        # glob reads in any image with the extension "image_dir/im_type/*"
        for file in glob.glob(os.path.join(image_dir, im_type, "*")):
            
            # Read in the image
            im = mpimg.imread(file)
            
            # Check if the image exists/if it's been correctly read-in
            if not im is None:
                # Append the image, and it's type (red, green, yellow) to the image list
                im_list.append((im, im_type))
    
    return im_list
````

````python
# Using the load_dataset function in helpers.py
# Load training data
IMAGE_LIST = load_dataset(training_dir)
````

## 2.0 Standardizing the Data

To build our image classifier we first need to **standardize our images**
        1. We standardize the input data (images)
             * Resize each image to a desired input size say 600x1100px   
        2. We standardize the output data (label)
             * Encode the label of the label of each corresponding images

---
### 2.1 Resizing input data

````python
# Standardize the input images

def standardize_input(image):
    
    standardize_image = cv2.resize(image, dsize=(1100, 600), interpolation=cv2.INTER_CUBIC)
    return standardize_image
````

---
### 2.2 Encoding output label

````python
# Standardize the output label
# We encode the label with 0 & 1 since it is a binary problem

def encode(label):
    if label == 'day':
        num_val = 1
    elif label == 'night':
        num_val = 0
    
    return num_val
````

We are going to construct a **`STANDARDIZED_LIST`** of input images and output labels.

This function takes in a list of image-label pairs and outputs a **standardized** list of resized images and numerical labels.

It uses the functions defined above to standardize the input and output.

---
### 2.3 Standardize the whole training data

````python
def standardize(IMAGE_LST):
    
    standardize_list = []
    
    for item in IMAGE_LST:
        image = item[0]
        label = item[1]
        
        standardized_image = standardize_input(image)
        
        encoded_lable = encode(label)
        
        standardize_list.append((standardized_image, encoded_lable))
        
    return standardize_list
````

````python
# Standardize all training data
STANDARDIZED_LIST = standardize(IMAGE_LIST)
````

## 3.0 Feature Extraction

Using the inference from section **1.3 Observation** we would create a clasifier based on the average brightness of the images to build a simple classifer

We'll create a feature that represents the brightness in an image. We'll be extracting the **average brightness** using HSV colorspace. Specifically, we'll use the V channel (a measure of brightness), add up the pixel values in the V channel, then divide that sum by the area of the image to get the average Value of the image.


---
### 3.1 convert from RGB to HSV color space

````python
image_idx = 45

img_ = STANDARDIZED_LIST[image_idx][0]
lbl_ = STANDARDIZED_LIST[image_idx][1]

# Convert to HSV
hsv = cv2.cvtColor(img_, cv2.COLOR_RGB2HSV)

# Print image label
print('Label :', str(lbl_))

# HSV channels
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

# Visualizing the HSV against the standardized data

f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,10))
ax1.set_title('Standardized image')
ax1.imshow(img_)
ax2.set_title('H channel')
ax2.imshow(h, cmap='gray')
ax3.set_title('S channel')
ax3.imshow(s, cmap='gray')
ax4.set_title('V channel')
ax4.imshow(v, cmap='gray')

````

---
### 3.2 Find the average brightness using the V channel

We would define a function that takes in a **standardized** RGB image and returns a feature (a single value) that represent the average level of brightness in the image. We'll use this value to classify the image as day or night.

````python
def average_brightness(rgb_img):
    
    # Convert image to HSV
    hsv = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2HSV)
    
    # Add up all the pixel values in the V channel
    brightness_sum = np.sum(hsv[:,:,2])
    
    # Calculate the average brightness using the area of the image
    area = 600*1100.0  # pixels
    
    
    avg = brightness_sum/area
    return avg
````

````python
night_brightness = []
day_brightness = []

for image in STANDARDIZED_LIST:
    
    if image[1] == 0:
        night_brightness.append(average_brightness(image[0]))
    elif image[1] == 1:
        day_brightness.append(average_brightness(image[0]))

avg_day_brightness = np.mean(day_brightness)
avg_night_brightness = np.mean(night_brightness)
min_day_brightness = np.min(day_brightness)
max_night_brightness = np.max(night_brightness)

print(f"Average of average day brightness from: {avg_day_brightness}")
print(f"Average of average night brightness: {avg_night_brightness}")
print(f"Minimum day brightness: {min_day_brightness}")
print(f"Maximum night brightness: {max_night_brightness}")

````

## 4.0 Classification and Visualizing Error

Now, we'll turn our average brightness feature into a classifier that takes in a standardized image and returns a `predicted_label` for that image. This `estimate_label` function should return a value: 0 or 1 (night or day, respectively).

---
### 4.1 Build a complete classifier

We are going athreshold that will seperate the day and night images by average brightness

````python
def label_estimate(rgb_image):
    
    avg = average_brightness(rgb_image)
    
    predicted_label = 0
    
    threshold = 100
    if avg > threshold:
        
        predicted_label = 1
        
    return predicted_label
````

### 4.2 Testing the classifier

Here is where we going to test the classification algorithm using our test set of data that we set aside at the beginning of the notebook!

Since we are using a pretty simple brightess feature, we may not expect this classifier to be 100% accurate. We'll aim for around 75-85% accuracy usin this one feature.


**Test dataset**

Below, we load in the test dataset, standardize it using the `standardize` function you defined above, and then **shuffle** it; this ensures that order will not play a role in testing accuracy.


````python
import random

# Load test data
TEST_IMAGE_LIST = load_dataset(test_dir)

# Standardize the test data
STANDARDIZED_TEST_LIST = standardize(TEST_IMAGE_LIST)

# Shuffle the standardized test data
random.shuffle(STANDARDIZED_TEST_LIST)
````

### 4.3 Determine the Accuracy

Compare the output of the classification algorithm with the true labels and determine the accuracy.

This code stores all the misclassified images, their predicted labels, and their true labels, in a list called `misclassified`.

````python
# Constructs a list of misclassified images given a list of test images and their labels
def get_misclassified_images(test_images):
    # Track misclassified images by placing them into a list
    misclassified_images_labels = []

    # Iterate through all the test images
    # Classify each image and compare to the true label
    for image in test_images:

        # Get true data
        im = image[0]
        true_label = image[1]

        # Get predicted label from your classifier
        predicted_label = label_estimate(im)

        # Compare true and predicted labels 
        if(predicted_label != true_label):
            # If these labels are not equal, the image has been misclassified
            misclassified_images_labels.append((im, predicted_label, true_label))
            
    # Return the list of misclassified [image, predicted_label, true_label] values
    return misclassified_images_labels
````

````python
# Find all misclassified images in a given test set
MISCLASSIFIED = get_misclassified_images(STANDARDIZED_TEST_LIST)

# Accuracy calculations
total = len(STANDARDIZED_TEST_LIST)
num_correct = total - len(MISCLASSIFIED)
accuracy = num_correct/total

print('Accuracy: ' + str(accuracy))
print("Number of misclassified images = " + str(len(MISCLASSIFIED)) +' out of '+ str(total))
````
